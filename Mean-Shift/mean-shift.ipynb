```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from numpy import linalg as la
import math
```

```python
x1d = np.load('x1d.npy')
x2d = np.load('x2d.npy')
```

```python
def eucli(x, y):
    v = x - y
    if len(v.shape) == 1:
        return np.sqrt(np.square(v))
    else:
        return np.sqrt(np.square(v).sum(axis = -1))

def flat(xs, x, h):
    xx = np.zeros(xs.shape[0]).astype('float')
    for i in range(len(xx)):
        if sum(eucli(xs[i], x)**2) <= h**2:
            xx[i] = 1
        else:
            xx[i] = 0
    return xx

def gau(xs, x, h):
    return np.exp((-(eucli(xs, x)**2))/(h**2))/(np.sqrt(2*math.pi*h))
```

```python
def mean_shift(xs, x, kernel, h = 1, max_iters = 100, tol = 1e-6):
    shift = [x]
    for i in range(max_iters):
        weight = kernel(xs, x, h)
        if sum(weight) == 0:
            break
        m = (xs.T * weight).sum(axis = -1) / sum(weight) - x
        x += m
        shift = np.vstack((shift, x))
        if np.dot(m, m) < tol:
            break
    return np.array(shift)
```

```python
x0 = [[1], [2.5], [-5]]
x1 = [[0,0], [-4, 10], [10, 10]]
```

```python
for x in x0:
    print(mean_shift(x1d, x, flat)[-1])
```

```python
for x in x0:
    print(mean_shift(x1d, x, gau)[-1])
```

```python
for x in x1:
    print(mean_shift(x2d, x, flat)[-1])
```

```python
for x in x1:
    print(mean_shift(x2d, x, gau)[-1])
```
